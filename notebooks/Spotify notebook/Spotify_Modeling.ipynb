```python
import pandas as pd


import seaborn as sns
import matplotlib.pyplot as plt

# models
from sklearn.svm import SVC
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier

# word vecotrizers
from sklearn.metrics import classification_report


import joblib
```

```python
X_train = joblib.load(r"../../Data\interim\Spotify\X_train.pkl")
X_test = joblib.load(r"../../Data\interim\Spotify\X_test.pkl")
y_train = joblib.load(r"../../Data\interim\Spotify\y_train.pkl")
y_test = joblib.load(r"../../Data\interim\Spotify\y_test.pkl")
```

# modeling
---

Best model: SVC_RBF with weighted F1 = 0.8654 <br>
RandomForestClassifier: weighted F1 = 0.8477 <br>
LGBMClassifier: weighted F1 = 0.8498 <br>

```python
best_model = SVC(kernel= 'rbf', probability=True)

best_model.fit(X_train,y_train)
```

```python
y_pred = best_model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
joblib.dump(best_model, "../../models/Spotify_model.pkl" )
```
